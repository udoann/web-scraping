### Download images from the page

```python
url_one = "https://keithgalli.github.io/web-scraping/"
images = soup.select("div.row div.column img")
images_url = [image['src'] for image in images]
images_url
```

```python
for i, url in enumerate(images_url):
    full_url = url_one + url
    img_data = requests.get(full_url).content
    with open('image_name'+ str(i) + '.png', 'wb') as handler:
        handler.write(img_data)
```

### Solve the Mystery Challenge

```python
files = soup.select('div.block a')
relative_files = [f['href'] for f in files]
relative_files
```

```python
for f in relative_files:
    full_url = url_one + f
    pages = requests.get(full_url)
    bs_pages = bs(pages.content)
    secret_word = bs_pages.find('p', attrs={'id':'secret-word'})
    m_secret_word = secret_word.string
    print(m_secret_word)
```
