```python
# create two more column goals per match
# Create minutes per match
# The dot Astypye i to convert the created column to a specific data type.
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df['MinutesPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df.head()
```

```python
# Find the total number of goals
total_goals = epl_df['Goals'].sum()
print(total_goals)
```

```python
# find number of penalty goals
penalty_goals = epl_df['Penalty_Goals'].sum()
print(penalty_goals)
```

```python
# Find total number of penalty Attempted
penalty_attempted = epl_df['Penalty_Attempted'].sum()
print(penalty_attempted)
```

```python
# PieChart of penalty missed vs penalty scored
penalty_missed = penalty_attempted - penalty_goals
x = [penalty_goals,penalty_missed]
labels = ['Penalty Scored', 'Penalty Missed']
color = sns.color_palette('Set2')
plt.pie(x, labels = labels, shadow = True, autopct = '%.0f%%', colors = color)
plt.figure(figsize = (13, 6))
plt.show()
```

```python
# Players by age group.
under_twenty = epl_df[epl_df['Age'] < 20]
twenty_to_five = epl_df[(epl_df['Age'] >=20) & (epl_df['Age'] <= 25)]
twenty_five_thirty = epl_df[(epl_df['Age'] > 25) &(epl_df['Age'] <= 30)]
above_thirty = epl_df[(epl_df['Age'] > 30)]
```

```python
x = np.array([under_twenty['Name'].count(), 
              twenty_to_five['Name'].count(), 
              twenty_five_thirty['Name'].count(),
              above_thirty['Name'].count()])
labels = ['<20','20-25','26-30','>30']
plt.title('chart of percentage of Different Age Group',{'fontsize':20, 'color':'indigo'})
plt.pie(x, labels=labels, colors=sns.color_palette('viridis'), autopct="%.1f%%")
plt.show()
```

```python
# Total Assist by Club
assists_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid", color_codes = False)
ax = sns.barplot(x='Club',y='Assists',data=assists_by_clubs.sort_values(by='Assists'),palette ='Set2')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('plot of clubs vs total Assists', fontsize= 20)
```

```python
# Top 10 Assits
tp_ten_assists =epl_df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
tp_ten_assists
```

```python
# Total Assist by Club
goals_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes = False)
ax = sns.barplot(x='Club',y='Goals',data=goals_by_clubs.sort_values(by='Goals'),palette ='rocket')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20,8)
plt.title('Plot of Clubs vs total Goals', fontsize= 20)
```
